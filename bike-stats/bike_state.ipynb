### What’s the average number of bikes at a given bike station?

```python
df_station = pd.read_csv('station.csv')
stations_avg =df_station[df_station['decommission_date'].isnull()]
stations_avg
stations=stations_avg['current_dockcount'].mean()
stations
```

### When a bike station is modified, is it more likely that it’ll lose bikes or gain bikes? How do you know?


```python
modified_station = df_station[['modification_date','install_dockcount','current_dockcount']].dropna()
lose_or_gain=modified_station['current_dockcount'].sum() - modified_station['install_dockcount'].sum()
lose_or_gain
```
